```python
import os
from skimage import io
import os.path
import matplotlib.pyplot as plt
from torch.utils.data import Dataset, DataLoader

IMAGE_DIR="./datasets/courier"

import numpy as np
import torchvision.transforms as transforms
import torch
from skimage import io

from torchvision.transforms import ToPILImage
to_img = ToPILImage()
```

```python
class LettersDataset(Dataset):
    """Letters dataset."""

    def __init__(self, root_dir, transform):
        """
        Args:
            root_dir (string): Directory with all the images.
        """
        self.root_dir = root_dir
        self.img_names = os.listdir(IMAGE_DIR)
        self.transform = transform

    def __len__(self):
        return len(self.img_names)

    def __getitem__(self, idx):
        img_name = os.path.join(self.root_dir,
                                self.img_names[idx])
        image = io.imread(img_name, as_gray = True)
                   
        image = torch.from_numpy(np.array([image], float))  
        image = image /255.0 - 0.5;
       
        sample = {'image': image}

        return sample
```

```python
transform = transforms.Compose(
    [#transforms.ToTensor(),
     transforms.Normalize((0.5, ), (0.5, ))])
    
dataset = LettersDataset(IMAGE_DIR, transform)
trainloader = DataLoader(dataset, batch_size=16, shuffle=False, num_workers=4)
```

```python
from torch.autograd import Variable
import torch.nn as nn
import torch.nn.functional as F

class Net(nn.Module):
    def __init__(self):
        self.first = True
        super(Net, self).__init__()
        
        # Set the number of (should be even because we are splitting it into two columns at some point
        self.INNER = 8
       
        ## Encoder
        self.conv1 = nn.Conv2d(1, 8, 3, stride=1)
        self.conv2 = nn.Conv2d(8, 16, 6, stride = 2)
        self.conv3 = nn.Conv2d(16, self.INNER, 13, stride = 1)
        
        self.fc_mean = nn.Linear(self.INNER, self.INNER)
        self.fc_logvar = nn.Linear(self.INNER, self.INNER)
        self.fc_latent = nn.Linear(self.INNER, self.INNER)
        
        ## Decoder        
        self.rconv3 = nn.ConvTranspose2d(self.INNER, 16, 13, stride = 1)
        self.rconv2 = nn.ConvTranspose2d(16, 8, 6, stride = 2)
        self.rconv1 = nn.ConvTranspose2d(8, 1, 3, stride = 1)
        
        
    def set_eval_mode(self):
        # Could be automated, but how ?
        self.bn1.eval()
        self.bn2.eval()
        self.bn3.eval()
        self.fc1.eval()
        self.bnl1.eval()
        self.bnlsm1.eval()
        self.bn3B.eval()
        self.bn2B.eval()
        self.bn1B.eval()
        
        
    def set_train_mode(self):
        # Could be automated, but how ?
        self.bn1.train()
        self.bn2.train()
        self.bn3.train()
        self.fc1.eval()
        self.bnl1.train()
        self.bnlsm1.train()
        self.bn3B.train()
        self.bn2B.train()
        self.bn1B.train()
        
    def encode(self, inputs):
        x = F.relu(self.conv1(inputs))
        # print("conv1 {}".format(x.size()))
        x = F.relu(self.conv2(x))
        # print("conv2 {}".format(x.size()))
        x = F.relu(self.conv3(x))
        # print("conv3 {}".format(x.size()))

        x = x.view(-1, self.INNER)

        return self.fc_mean(x), self.fc_logvar(x)
        
    def decode(self, z):
        x = self.fc_latent(z)

        x = x.view(-1, self.INNER, 16, 1)

        x = F.relu(self.rconv1(x))
        # print("rconv1 {}".format(x.size()))
        x = F.relu(self.rconv2(x))
        # print("rconv2 {}".format(x.size()))
        x = F.relu(self.rconv3(x))
        # print("rconv3 {}".format(x.size()))
        
        return x
    
    def forward(self, x, deterministic=False):
        mean, logvar = self.encode(x)
        latent = self.reparameterize(mean, logvar)

        if deterministic:
            latent = mean

        encoded = self.decode(latent)

        return latent, encoded, mean, logvar
    
    def reparameterize(self, mean, logvar):
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        return eps * std + mean
```

```python
net = Net()
```

```python
import torch.optim as optim

criterion = nn.MSELoss()
optimizer = optim.SGD(net.parameters(), lr=0.1, momentum=0.5)


def imshow_pair(image_pair, titles=('', ''), figsize=(10, 5), **kwargs):
    fig, axes = plt.subplots(ncols=2, figsize=figsize)
    for ax, img, label in zip(axes.ravel(), image_pair, titles):
        ax.imshow(img, **kwargs)
        ax.set_title(label)
        

from torchvision.transforms import ToPILImage
to_img = ToPILImage()
net.first = True
net.set_train_mode()
for epoch in range(10000):  # loop over the dataset multiple times

    running_loss = 0.0
    for i, data in enumerate(trainloader, 0):
        # get the inputs
        image = data["image"]
        image_data = ((image[1,:]+0.5)*255.0).type(torch.IntTensor)
#         print(image_data)
#         plt.imshow(transforms.ToPILImage()(image_data))
#         plt.show() 
        
#        break
        # This is an encoder decoder so output and input are identical
        # wrap them in Variable
    
#        if i == 4:
#            i = to_img(image[0].normal_().float())
#            plt.imshow(i)
#            plt.show()
#            break
    
        inputs, labels = Variable(image).float(), Variable(image).float()
        
#         print(inputs, labels)

        # zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        outputs = net(inputs)
        
        if epoch %40 == 0:           
            o = outputs.data.numpy()
#            print("min", o.min(), "max", o.max())
#            print(o.shape)
            #i = to_img(o)
            #plt.imshow(o)
            #plt.show()
    
        
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        # print statistics
#         print("loss: ",loss)
        running_loss += loss
#         for j in range(loss.data.shape[0]):
#             running_loss += loss.data[j]
    if epoch % 100 == 0:    # print every 200 epoch
       imshow_pair((inputs.data[1][0], outputs.data[1][0]))
       plt.show()
       net.first = True
       print('[%d, %5d] loss: %.4f' %
           (epoch + 1, i + 1, running_loss))
       running_loss = 0.0

print('Finished Training')
```

```python
from sklearn import manifold
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
# https://github.com/despoisj/LatentSpaceVisualization/blob/master/visuals.py

# Show dataset images with T-sne projection of latent space encoding
def computeTSNEProjectionOfLatentSpace(X, outputs, display=True):

    # Compute t-SNE embedding of latent space
    print("Computing t-SNE embedding...")
    X_tsne = manifold.TSNE(n_components=2).fit_transform(X.reshape(-1,8))
    # Plot images according to t-sne embedding
    if display:
        print("Plotting t-SNE visualization...")
        fig, ax = plt.subplots()
        imscatter(X_tsne[:, 0], X_tsne[:, 1], imageData=outputs, ax=ax, zoom=1)
        plt.show()
    else:
        return X_tsne
    
# Scatter with images instead of points
def imscatter(x, y, ax, imageData, zoom):
    images = []
    for i in range(len(x)):
        x0, y0 = x[i], y[i]
        # Convert to image
        # Note: OpenCV uses BGR and plt uses RGB
        image = OffsetImage(imageData[i,:,:], zoom=zoom)
        ab = AnnotationBbox(image, (x0, y0), xycoords='data', frameon=False)
        images.append(ax.add_artist(ab))
    
    ax.update_datalim(np.column_stack([x, y]))
    ax.autoscale()

displayloader = DataLoader(dataset, batch_size=1, shuffle=False, num_workers=1)
net.set_eval_mode()
X = torch.zeros(16, 4, 2)
outputs = torch.zeros(16, 32, 32)

for i, data in enumerate(displayloader, 0):
    image = data["image"]
    inputs, labels = Variable(image).float(), Variable(image).float()    
    X[i,:, :] = net.inner_repr.data[0]
    output = net(inputs)
    outputs[i,:, :] = output.data[0][0].squeeze()

computeTSNEProjectionOfLatentSpace(X, outputs)

    


```
